## EDA

```python
print(customers_df.describe(include='all'))
print(products_df.describe(include='all'))
print(transactions_df.describe(include='all'))
```

## Merging the Complete Data

```python
# Merge transactions with products
transactions_products = pd.merge(transactions_df, products_df, on='ProductID', how='inner')

# Merge the result with customers
full_data = pd.merge(transactions_products, customers_df, on='CustomerID', how='inner')

# Check the resulting dataframe
print(full_data.head())
```

## Revenue Analysis

```python
# Total revenue
total_revenue = full_data['TotalValue'].sum()
print(f"Total Revenue: ${total_revenue}")

# Revenue by Product
revenue_by_product = full_data.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False)
print(revenue_by_product.head())

# Revenue by Region
revenue_by_region = full_data.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)
print(revenue_by_region)
```

## Top 10 High Value Customers

```python
customer_revenue = full_data.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False)
print(customer_revenue.head(10)) 
```

## Revenue by Category

```python
category_revenue = full_data.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)
print(category_revenue)
```

## Trends over Time

```python
# Convert 'TransactionDate' to datetime if not already done
full_data['TransactionDate'] = pd.to_datetime(full_data['TransactionDate'])

# Monthly revenue trends
full_data['MonthYear'] = full_data['TransactionDate'].dt.to_period('M')
revenue_trends = full_data.groupby('MonthYear')['TotalValue'].sum()
print(revenue_trends)

# Plot revenue trends
revenue_trends.plot(kind='line', title='Monthly Revenue Trends', figsize=(10, 6))
plt.xlabel('Month-Year')
plt.ylabel('Revenue')
plt.show()
```

## Revenue to Region Graph

```python
# Bar chart for revenue by region
revenue_by_region.plot(kind='bar', title='Revenue by Region', figsize=(8, 6))
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

## Most Purchased Categories by Value and by Number of Units

```python
# Revenue by category
category_revenue = full_data.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)

# Number of units purchased by category
category_units = full_data.groupby('Category')['Quantity'].sum().sort_values(ascending=False)

# Visualize revenue by category
plt.figure(figsize=(10, 6))
category_revenue.plot(kind='bar', title='Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()

# Visualize units purchased by category
plt.figure(figsize=(10, 6))
category_units.plot(kind='bar', title='Units Purchased by Category')
plt.xlabel('Category')
plt.ylabel('Total Units')
plt.xticks(rotation=45)
plt.show()
```

## Average Purchase per Customer

```python
# Average purchase per customer (Total value spent / Number of customers)
avg_purchase_per_customer = full_data.groupby('CustomerID')['TotalValue'].sum().mean()

print(f"Average Purchase per Customer: ${avg_purchase_per_customer:.2f}")
```

## Average Purchase per Customer per Month

```python
# Total revenue per customer per month
monthly_purchase_per_customer = full_data.groupby(['CustomerID', 'MonthYear'])['TotalValue'].sum().unstack()

# Average purchase per customer for each month
avg_purchase_per_month = monthly_purchase_per_customer.mean(axis=0)

# Plotting the average purchase per customer for each month
plt.figure(figsize=(10, 6))
avg_purchase_per_month.plot(kind='line', marker='o', title='Average Purchase per Customer per Month')
plt.xlabel('Month-Year')
plt.ylabel('Average Purchase per Customer ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Sales Distribution per Region

```python
# Total sales value by region
region_revenue = full_data.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)

# Plotting sales by region
plt.figure(figsize=(10, 6))
region_revenue.plot(kind='pie', autopct='%1.1f%%', startangle=90, title='Sales Distribution by Region')
plt.ylabel('')
plt.show()
```

## Repeat Customers

```python
# Count the number of transactions for each customer
customer_transactions_count = full_data.groupby('CustomerID')['TransactionID'].nunique()

# Customers who made more than 1 purchase (repeat customers)
repeat_customers = customer_transactions_count[customer_transactions_count > 1]

# Plotting the distribution of repeat vs. non-repeat customers
plt.figure(figsize=(8, 6))
repeat_customers.value_counts().plot(kind='bar', color='skyblue', title='Repeat vs Non-Repeat Customers')
plt.xlabel('Number of Purchases')
plt.ylabel('Customer Count')
plt.xticks(rotation=0)
plt.show()
```

## Correlation Between Product Price and Total Sales

```python
# Calculate the correlation between product price (Price_x) and total sales value (TotalValue)
correlation = full_data[['Price_x', 'TotalValue']].corr()

# Plotting the correlation matrix
plt.figure(figsize=(6, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation between Product Price and Total Sales Value')
plt.show()
```

## Most Active Customers

```python
# Count the number of transactions per customer
active_customers = full_data.groupby('CustomerID')['TransactionID'].nunique().sort_values(ascending=False)

# Plotting the most active customers
plt.figure(figsize=(12, 6))
active_customers.head(10).plot(kind='bar', color='orange', alpha=0.7, title='Most Active Customers')
plt.xlabel('Customer ID')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.show()
```

## Purchase Trends per Region

```python
# Convert 'MonthYear' to string type for proper plotting
full_data['MonthYear'] = full_data['MonthYear'].astype(str)

# Calculate total sales and quantity per region and month
region_monthly_sales = full_data.groupby(['Region', 'MonthYear'])[['TotalValue', 'Quantity']].sum().reset_index()

# Pivot the table to make plotting easier
region_sales_pivot = region_monthly_sales.pivot(index='MonthYear', columns='Region', values='TotalValue')

# Plot the total sales trends per region
plt.figure(figsize=(12, 6))
for region in region_sales_pivot.columns:
    plt.plot(region_sales_pivot.index, region_sales_pivot[region], label=region)

plt.title('Total Sales Trends Per Region')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```
